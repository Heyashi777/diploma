```python
fig = go.Figure()
fig.add_trace(go.Scatter(y=data_portfel["future_income"], x=data_portfel["period"], mode="lines", name="Прогозируемая доходность, %"))
fig.add_trace(go.Scatter(y=data_portfel["fact_income"], x=data_portfel["period"], mode="lines", name="Реальная доходность, %"))
for i in range(len(data_portfel) - 1):
    x0 = data_portfel["period"].iloc[i]
    x1 = data_portfel["period"].iloc[i + 1]

    y0_fut = data_portfel["future_income"].iloc[i]
    y1_fut = data_portfel["future_income"].iloc[i + 1]

    y0_fact = data_portfel["fact_income"].iloc[i]
    y1_fact = data_portfel["fact_income"].iloc[i + 1]

    x_fill = [x0, x1, x1, x0]
    y_fill = [y0_fut, y1_fut, y1_fact, y0_fact]

    color = "rgba(255, 0, 0, 0.2)" if y0_fut > y0_fact else "rgba(0, 200, 0, 0.2)"

    fig.add_trace(go.Scatter(
        x=x_fill,
        y=y_fill,
        fill="toself",
        mode="lines",
        line=dict(width=0),
        fillcolor=color,
        showlegend=False
    ))
fig.update_layout(
        title="Сравнение прогнозируемой и фактическоой доходностей",
        xaxis_title="период",
        yaxis_title="%",
        plot_bgcolor="white",
        xaxis=dict(showgrid=True),
        yaxis=dict(showgrid=True),
    )
fig.show()
```

```python
def cumulative_log_return(df: pd.DataFrame) -> pd.DataFrame:
    log_returns = np.log(df / df.shift(1))  # Логарифмическая доходность
    cumulative_log_returns = log_returns.cumsum()  # Кумулятивная логарифмическая доходность
    return cumulative_log_returns

data_moex = pd.read_csv("data/MOEX.csv", index_col="TRADEDATE", parse_dates=["TRADEDATE"])
# data_moex = data_moex['return'][data_moex.index  >= "2014-01-01 00:00:00"]

moex = data_moex
# moex = moex.set_index("TRADEDATE")
moex = moex[moex.index >= '2014-07-01 00:00:00']
moex = moex[moex.index < '2025-01-01 00:00:00']
periods = generate_half_year_dates(moex)
moex['income'] = cumulative_log_return(moex['OPEN'])

income = []
for per in periods:
    stop = per + pd.DateOffset(months=6)
    period_data = moex[(moex.index >= per) & (moex.index < stop)].copy()
    # print(per,  period_data['income'].mean())
    income.append(1+period_data['income'].mean())

sub = data_portfel
sub['fact_proc'] = 1 + sub["fact_income"]/200
sub["fact_cum_proc"] = sub["fact_proc"].cumprod()
sub["fact_cum_percent"] = (sub["fact_proc"].cumprod() - 1) * 100

fig = go.Figure()
fig.add_trace(go.Scatter(y=income, x=periods, mode="lines+markers", name="MOEX"))
fig.add_trace(go.Scatter(y=sub["fact_cum_proc"], x=sub["period"], mode="lines+markers", name="портфель"))
fig.add_hline(y=1.5, line_color='black', line_width=2, line_dash='dot')
fig.add_hline(y=2, line_color='black', line_width=2, line_dash='dot')
fig.add_vline(x='2019-01-01', line_color='black', line_width=2, line_dash='dot')
fig.add_vline(x='2024-01-01', line_color='black', line_width=2, line_dash='dot')
for i in range(len(sub) - 1):
    x0 = sub["period"].iloc[i]
    x1 = sub["period"].iloc[i + 1]

    y0_fut = sub["fact_cum_proc"].iloc[i]
    y1_fut = sub["fact_cum_proc"].iloc[i + 1]

    y0_fact = income[i]
    y1_fact = income[i + 1]

    x_fill = [x0, x1, x1, x0]
    y_fill = [y0_fut, y1_fut, y1_fact, y0_fact]

    color = "rgba(255, 0, 0, 0.2)" if y0_fut > y0_fact else "rgba(0, 200, 0, 0.2)"

    fig.add_trace(go.Scatter(
        x=x_fill,
        y=y_fill,
        fill="toself",
        mode="lines",
        line=dict(width=0),
        fillcolor=color,
        showlegend=False
    ))
fig.update_layout(
        title="Сравнение прогнозируемой и фактическоой доходностей",
        xaxis_title="период",
        yaxis_title="%",
        plot_bgcolor="white",
        xaxis=dict(showgrid=True),
        yaxis=dict(showgrid=True),
    )
fig.update_layout(
        title="Сравнение доходности индекса MOEX и портфеля",
        xaxis_title="период",
        yaxis_title="%",
        plot_bgcolor="white",
        xaxis=dict(showgrid=True),
        yaxis=dict(showgrid=True),
    )
fig.show()
```
