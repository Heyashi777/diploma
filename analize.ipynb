```python
import pandas as pd
from utils import generate_half_year_dates
```

```python
data_moex = pd.read_csv("data/returns_of_stok.csv", index_col="TRADEDATE", parse_dates=["TRADEDATE"])
data_moex = data_moex['MOEX'][data_moex.index  >= "2014-01-01 00:00:00"]
```

```python
data_por = pd.read_csv('data/portfels.csv')
```

```python
data_por
```

```python
periods = generate_half_year_dates(data_moex)
periods
```

```python
res = []
for per in periods:
    stop = per + pd.DateOffset(months=6)
    period_data = data_moex[(data_moex.index >= per) & (data_moex.index < stop)].copy()
    print(f"{per}-{stop}:  {period_data.mean()*100*2:.2f}  ")
    res.append(period_data.mean()*100*2)
```

```python
len(res), len(data_por)
```

```python
data_por['MOEX'] = res[:-1]
data_por
```
