```python
import requests
import pandas as pd
import time  # полезно на случай пауз между запросами
from tqdm import tqdm
base_url = "https://iss.moex.com/iss/history/engines/stock/markets/index/boards/SNDX/securities/IMOEX.json"
step = 100  # Максимум строк за один запрос
all_data = []
start = 0

while tqdm(True):
    url = f"{base_url}?start={start}&interval=24"
    response = requests.get(url)
    data = response.json()

    # Достаем таблицу history.data и её заголовки
    columns = data['history']['columns']
    rows = data['history']['data']

    if not rows:
        break  # Прекращаем, если данных больше нет

    df_page = pd.DataFrame(rows, columns=columns)
    all_data.append(df_page)

    start += step
    time.sleep(0.2)  # пауза, чтобы не получить блокировку

# Объединяем все страницы в один DataFrame
df_all = pd.concat(all_data, ignore_index=True)

# Преобразуем дату в формат datetime
df_all['TRADEDATE'] = pd.to_datetime(df_all['TRADEDATE'])

# Можно отфильтровать нужные столбцы:
df_cleaned = df_all[['TRADEDATE', 'OPEN', 'VOLUME', 'VALUE']]
df_cleaned
```

```python
df_cleaned["TRADEDATE"] = pd.to_datetime(df_cleaned["TRADEDATE"], errors="coerce")
df_cleaned = df_cleaned.set_index("TRADEDATE")
df_cleaned.fillna(method="ffill", inplace=True)
df_cleaned
```

```python
df_cleaned = df_cleaned[df_cleaned.index >= "2013-01-01 00:00:00"]
df_cleaned
```

```python
df_cleaned.to_csv('data/MOEX.csv')
```
